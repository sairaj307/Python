**Stack**: A stack is a linear data structure that follows the **Last In, First Out (LIFO)** principle, meaning the element added last is removed first.

It is widely used in programming for tasks such as expression evaluation, backtracking, and function call management.

```python
stack = []
def push():
    stack.append(input("Enter element"))
    print(stack)
def pop():
    if not stack:
        print("Stack is empty")
    else:
        print("removed element is: ", stack.pop())
        print(stack)
while True:
    print("Select the operation 1.push 2.pop 3.quit")
    choice = int(input())
    if choice == 1:
        push()
    elif choice == 2:
        pop()
    elif choice == 3:
        break
    else:
        print("Enter the correct operation")
```

**Queue**: It is a Linear data structure where elements are inserted from one end and removed from another end.

It follows First In First out or Last in Last out Principal

Operations performed by the queue

**Enqueue**: Adds an item to the queue. If the queue is full, then it is said to be an Overflow condition - Time Complexity : O(1)

**Dequeue**: Removes an item from the queue. The items are popped in the same order in which they are pushed. If the queue is empty, then it is said to be an Underflow condition - Time Complexity : O(1)

**Front**: Get the front item from queue - Time Complexity : O(1)

**Rear**: Get the last item from queue - Time Complexity : O(1)


```python
#Queue implementation using list
queue = []
def enqueue():
    queue.append(input("Enter element: "))
    print(queue)
def dequeue():
    if not queue:
        print("Queue is empty")
    else:
        print("Removed element: ", queue[0])
        queue.pop(0)
        print("Queue after removing: ", queue)
def front():
    print("Front element: ", queue[0])
def rear():
    print("Rear element: ", queue[-1])
```

Each node contains data field find a link or reference
Node elements of the linked list
Each node has two fields:
1. Data field
2. Reference to other node or null

Head starting node or first node

Last node is called as end

It has non contiguous memory location

```python
# SINGLY LINKED LIST
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
class LinkedList:
  def __init__(self):
    self.head = None

  def printLinkedList(self):
    if self.head is None:
      print("Linked list is empty")
      return
    current_node = self.head
    while current_node:
      print(current_node.data)
      current_node = current_node.next

  def insertAtBeginning(self, new_data):
    new_node = Node(new_data)
    new_node.next = self.head
    self.head = new_node
    self.printLinkedList()

  def insertAtEnd(self, new_data):
    new_node = Node(new_data)
    end = None
    temp = self.head
    while temp.next is not None:
      temp = temp.next
    end = temp
    end.next = new_node
    self.printLinkedList()

  def insertAfter(self, previous_node_data, data):
    node = Node(data)
    temp = self.head
    while temp.data != previous_node_data:
        if temp.next is None:
            print(previous_node_data, " not found")
            return
        temp = temp.next
    node.next = temp.next
    temp.next = node
    self.printLinkedList()

  def insertBefore(self, node_data, data):
    node = Node(data)
    temp = self.head
    while temp.next is not None and temp.next.data != node_data:
        temp = temp.next
    if temp.next is None:
            print(node_data, " not found")
            return
    node.next = temp.next
    temp.next = node
    self.printLinkedList()

  def deleteHead(self):
    if self.head is None:
      print("Linked list is empty")
      return
    temp = self.head
    self.head = temp.next
    temp = None
    self.printLinkedList()

  def delete(self, givenData):
    temp = self.head
    if temp is not None and temp.data == givenData:
        self.deleteHead()
    else:
        while temp.next is not None and temp.next.data != givenData:
            temp = temp.next
        if temp.next is None:
            print("Element not found")
            return
        temp2 = temp.next
        temp.next = temp.next.next
        temp2.next = None
    self.printLinkedList()

  def deleteLastNode(self):
      temp = self.head
      if self.head is None:
          print("Empty linked list")
      elif temp.next is None:
          self.deleteHead()
      else:
          while temp.next.next is not None:
              temp = temp.next
          temp.next = None
```

```python
# Doubly Linked List
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None

  def printLinkedList(self):
    if self.head is None:
      print("Linked list is empty")
      return
    current_node = self.head
    while current_node:
      print(current_node.data)
      current_node = current_node.next

  def insertAtBeginning(self, new_data):
    new_node = Node(new_data)
    new_node.next = self.head
    if self.head:
      self.head.prev = new_node
    self.head = new_node
    self.printLinkedList()

  def insertAtEnd(self, data):
      newNode = Node(data)
      if self.head:
          temp = self.head
          while temp.next:
              temp = temp.next
          temp.next = newNode
          newNode.prev = temp
      else:
          self.head = newNode
      self.printLinkedList()

  def insertAfter(self, expectedNodeData, data):
        newNode = Node(data)
        temp = self.head
        if temp:
            while temp.data != expectedNodeData:
                temp = temp.next
                if not temp:
                    print(expectedNodeData, "not found in linked list")
                    return
            if not temp.next:
                self.insertAtEnd(data)
                return
            else:
                nextNode = temp.next
                temp.next = newNode
                newNode.prev = temp
                newNode.next = nextNode
                nextNode.prev = newNode
        else:
            print("linked list is empty")
            return
        self.printLinkedList()

  def insertBefore(self, expectedNodeData, data):
        newNode = Node(data)
        temp = self.head
        if temp:
            while temp.data != expectedNodeData:
                temp = temp.next
                if not temp:
                    print(expectedNodeData, "not found in linked list")
                    return
            if not temp.prev:
                self.insertAtBeginning(data)
                return
            else:
                newNode.next = temp
                newNode.prev = temp.prev
                temp.prev.next = newNode
                temp.prev = newNode
        else:
            print("linked list is empty")
            return
        self.printLinkedList()

  def deleteHead(self):
      temp = self.head
      if self.head:
          if self.head.next:
              self.head = self.head.next
              self.head.prev = None
              temp.next = None
          else:
              self.head = None
      else:
          print("Linked List is empty")
          return
      self.printLinkedList()

  def deleteLastNode(self):
        temp = self.head
        if temp is None:
            print("Linked List is empty")
        elif temp.next is None:
            self.head = None
        else:
            while temp.next:
                temp = temp.next
            temp.prev.next = None
            temp.prev = None
        self.printLinkedList()

  def delete(self, data):
        temp = self.head
        if temp is None:
            print("Linked List is empty")
        elif temp.next is None:
            self.head = None
            self.printLinkedList()
        else:
            while temp.data != data:
                temp = temp.next
                if temp is None:
                    print("Element not found")
                    return
            if temp.prev is None:
                self.deleteHead()
            elif temp.next is None:
                self.deleteLastNode()
            else:
                temp.prev.next = temp.next
                temp.next.prev = temp.prev
                temp.prev = None
                temp.next = None
                self.printLinkedList()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyCircularLinkedList:
    def __init__(self):
        self.head = None

    def insertFront(self, data):
        node = Node(data)
        if self.head is None:
            self.head = node
            node.next = node.prev = node
            return
        else:
            node.next = self.head
            node.prev = self.head.prev
            self.head.prev = node.prev.next = node
            self.head = node

    def insertAtEnd(self, data):
        node = Node(data)
        if self.head is None:
            self.head = node
            node.next = node.prev = node
            return
        else:
            node.next = self.head
            node.prev = self.head.prev
            self.head.prev = node
            node.prev.next = node

    def insertAfter(self, givenData, newData):
        node = Node(newData)
        if self.head == None:
            print("list is empty")
            return
        temp = self.head
        while temp.data != givenData:
            temp = temp.next
            if temp == self.head:
                print(givenData, " not present")
                return
        node.prev = temp
        node.next = temp.next
        temp.next = node.next.prev = node

    def deleteTail(self):
        if self.head is None:
            print("List is empty")
            return
        elif self.head.next == self.head.prev == self.head:
            self.head = None
            return
        else:
            self.head.prev = self.head.prev.prev
            self.head.prev.next = self.head

    def deleteHead(self):
        if self.head is None:
            print("List is empty")
            return
        elif self.head.next == self.head.prev == self.head:
            self.head = None
            return
        else:
            self.head.next.prev = self.head.prev
            self.head.prev.next = self.head.next
            self.head = self.head.next

    def deleteNode(self, data):
        if self.head is None:
            print("List is empty")
            return
        else:
            temp = self.head
            while True:
                if temp.data == data:
                    if temp == self.head:
                        self.deleteHead()
                        return
                    temp.prev.next = temp.next
                    temp.next.prev = temp.prev
                    temp.prev = temp.next = temp.data = None
                    return
                temp = temp.next
                if temp == self.head:
                    print(data, " not found in list")
                    return

    def printList(self):
        if self.head is None:
            print("empty list")
            return
        elif self.head.next == self.head:
            print(self.head.data)
            return
        else:
            current = self.head
            while True:
                print(current.data)
                current = current.next
                if current == self.head:
                    return
```

**Tree**: Tree is is a non-linear data structure in which a collection of elements known as nodes are connected to each other via edges such that there exists exactly one path between any two nodes.

1. **Node**: Individual elements of tree is called node.
Every node contains data and value and link to the other node
2. **Root Node**: Top most node is called as root node.
Origin of the tree data structure.
3. **Parent**: A node which has branch from it to any other node is parernt node.
4. **Child**: A node which has link from it's parent is called as child node.
5. **Sibling Nodes**: Nodes with same parent is called as Sibling Nodes.
6. **Leaf/External/Terminal Node**: Node which does not have child.
7. **Degree of a Node**: Total number of children of that node.
8. **Degree of a Tree**: Highest degree of a node among all the nodes in a tree.
9. **Level of Node**: Distance of node from root.
10. **Height of a Node**: Longest path (in edges) from that node to a leaf.
11. **Height of a tree**: Height of root node/Longest path (in edges) from root node to leaf node.
11. **Depth of a Node**: The number of edges from the root node to that specific node.
12. **Depth of a Tree**: The longest path from the root node to any leaf node (measured in edges).

**Characteristics of Tree**:
1. A Tree can contain only one root node.
2. There are N-1 edges for N nodes.
3. Every children have one parent but Parent can have multiple children.
4. Tree is a recursive data structure: Tree contains tree within them, Tree contains sub trees.
5. Degree of leaf node is 0.
6. Height of a leaf node is always 0.
7. Depth of root node is always 0.

Types of Binary Tree:
1.  Full Binary Tree: A tree in which every node has 0 or 2 children  no node has only one child.
2. Perfect Binary Tree: A Tree is both full and all leaf nodes are at the same level.

```python
def max_product_subarray(arr):
    max_prod = min_prod = result = arr[0]

    for num in arr[1:]:
        if num < 0:
            max_prod, min_prod = min_prod, max_prod

        max_prod = max(num, max_prod * num)
        min_prod = min(num, min_prod * num)

        result = max(result, max_prod)

    return result
```

```python
def pairSum(arr, target):
    i = 0
    j = len(arr)-1
    vect = []
    while(i<j):
        pair_sum = arr[i] + arr[j]
        if pair_sum < target:
            i+=1
        elif pair_sum > target:
            j-=1
        else:
            vect.append(i)
            vect.append(j)
            return vect
    return vect
```

```python
def majority(arr):
    arr.sort()
    freq = 1
    maj = arr[0]
    n = int(len(arr)/2) #4
    #[2, 3, 3, 3, 3, 3, 5, 8, 11]
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            freq += 1
        else:
            freq = 1
            maj = arr[i]
        if freq > n:
            return maj
```

```python
def sortList(arr):
  n = len(arr)
  for j in range(n-1):
      for i in range(n-j-1):
        if arr[i] > arr[i+1]:
            arr[i] , arr[i+1] = arr[i+1], arr[i]
```

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        print("START strStr()")
        yes = True
        print("Initialized yes =", yes)

        print("LENGTHS:", len(haystack), len(needle))

        print("Checking: if needle in haystack:", needle, haystack)
        if needle in haystack:
            print("NEEDLE FOUND IN HAYSTACK (initial test passed)")

            for i in range(len(haystack)):
                print("\n--- OUTER LOOP i =", i, " ---")
                print("Comparing haystack[i] with needle[0]:",
                      haystack[i], needle[0])

                if haystack[i] is needle[0]:
                    print("MATCH for first character at i =", i)

                    print("Entering INNER LOOP for needle match")

                    for k in range(len(needle)):
                        print(f"   Checking needle[{k}] vs haystack[{i+k}] :",
                              needle[k], haystack[i+k])

                        if needle[k] is not haystack[i+k]:
                            print("    MISMATCH at k =", k)
                            yes = False
                            print("   Setting yes =", yes)
                        else:
                            print("    MATCH at k =", k)

                    print("After INNER loop, yes =", yes)

                    if yes:
                        print("FULL MATCH FOUND at index", i)
                        return i
                else:
                    yes = True
                    print("Characters did NOT match for first character")

        print("No match found, returning nothing (None)")
        return -1
```

```python
from typing import List

class Solution:
    def maxArea(self, height: List[int]) -> int:
        #[1,8,6,2,5,4,8,3,7]
        maxArea = height[0]
        for i in range(len(height)-1):
            for j in range(i+1, len(height)):
                area = min(height[i], height[j])
                maxArea = max(maxArea, area*(j-i))
        return maxArea
```

```python
Solution().maxArea([1,8,6,2,5,4,8,3,7])
```

```python
def reverse(arr):
    i = 0
    j = len(arr)-1
    while i<j:
        arr[i], arr[j] = arr[j], arr[i]
        i+=1
        j-=1
    return arr
```

```python
def palindrome(s):
    isPalindrome = True
    i = 0
    j = len(s)-1
    while i<j:
        if s[i] != s[j]:
            isPalindrome = False
            break
        i+=1
        j-=1
    return isPalindrome        
```

```python
def productOfArrayExceptSelf(nums):
  n = len(nums)
  result = [1] * n
  container = 1
  for i in range(n):
      result[i] = container
      container *= nums[i]
  container = 1
  for i in range(n-1, -1, -1):
      result[i] *= container
      container *= nums[i]
  return result
```
